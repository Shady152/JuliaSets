```python
#ENPH316 Problem Set 1

from turtle import color
import numpy as np
import matplotlib.pyplot as plt

randNum = 500000 #number of random points <----|THE LOWER THE FASTER
divDepth = 100 #depth of divergence check <----|ANY DEEPER SEEMS IRRELEVANT FOR A SCATTER

datam = np.zeros(randNum, dtype = complex) #make an array to store the random points in the set

# |--------> randc will be used as the set of complex points to get the set <--------|

#Returns true or false depending if the random number is in the set by checking divergence
def mandDivCheck(z,c):
    for j in range(divDepth):
            z = np.exp(z) + c
            if(abs(z) > 100): return False
    return True

#Checks the random points to see if they belong in the set
def mandelbrot(z,c):
    k = 0
    for i in range(randNum):
        if(mandDivCheck(z,c[i])): 
            datam[k] = c[i]
            k += 1
    return datam

```

```python

randx = 200*(np.random.rand(randNum)-0.5) #generate array of random x points
randy = 200*(np.random.rand(randNum)-0.5) #generate array of random y points
randc = randx + randy*1j #generate array of random complex points

mand = mandelbrot(0,randc) #The mandelbrot starts at z = 0

#Mandelbrot Set Scatter Plot
plt.plot(np.real(mand), np.imag(mand), 'o', color = 'blue', ms = 0.5)
plt.title("Question 4 Mandelbrot Plot")
plt.xlabel("Re")
plt.xlim(-110,100)
plt.ylabel("Im")
#plt.savefig('mandelbrotQ4.png', dpi=800) #<-----------|UNCOMMENT TO SAVE PLOTS
plt.show()
```

```python

#The zoomed in Mandelbrot Set required the mult factor to be reduced since it acts like an axis size modifier
xzoom2 = 2*(np.random.rand(randNum)-0.5)-0.1
yzoom2 = 1*(np.random.rand(randNum)-0.5)
czoom2 = xzoom2 + yzoom2*1j

zoom2 = mandelbrot(0,czoom2)

#Zoomed in #Mandelbrot Set Scatter Plot
plt.plot(np.real(zoom2), np.imag(zoom2), 'o', color = 'green', ms = 0.5)
plt.title("Question 4 Zoomed Mandelbrot Plot 2")
plt.xlabel("Re")
plt.ylabel("Im")
#plt.savefig('mandelbrotQ4Zoom2.png', dpi=800) #<-----------|UNCOMMENT TO SAVE PLOTS
plt.show()
```

```python

#The zoomed in Mandelbrot Set required the mult factor to be reduced since it acts like an axis size modifier
randxzoom = 20*(np.random.rand(randNum)-0.5)-1
randyzoom = 20*(np.random.rand(randNum)-0.5)
randczoom = randxzoom + randyzoom*1j

mandZoom = mandelbrot(0,randczoom)

#Zoomed in #Mandelbrot Set Scatter Plot
plt.plot(np.real(mandZoom), np.imag(mandZoom), 'o', color = 'green', ms = 0.5)
plt.title("Question 4 Zoomed Mandelbrot Plot")
plt.xlabel("Re")
plt.ylabel("Im")
#plt.savefig('mandelbrotQ4Zoom.png', dpi=800) #<-----------|UNCOMMENT TO SAVE PLOTS
plt.show()
```

```python
#------------------------------------------------------------------------ Q5 Julia

randxj = 4*(np.random.rand(randNum)-0.5) #generate array of random x points
randyj = 4*(np.random.rand(randNum)-0.5) #generate array of random y points
randcj = randxj + randyj*1j #generate array of random complex points

dataj = np.zeros(randNum, dtype = complex) #make an array to store the random points in the set
numc = -0.83 + 0.18j #complex number for set

#Returns true or false depending if the random number is in the set by checking divergence
def julDivCheck(z,c):
    for j in range(divDepth):
            z = z**2 + c
            if(abs(z) > 2): return False
    return True

#Checks the random points to see if they belong in the set
def julia(z,c):
    k = 0
    for i in range(randNum):
        if(julDivCheck(z[i],c)): 
            dataj[k] = randcj[i]
            k += 1
    return dataj

julia = julia(randcj,numc)

#Julia Set Scatter Plot
plt.plot(np.real(julia), np.imag(julia), 'o', color = 'red', ms = 0.5)
plt.title("Question 5 Julia Set Plot")
plt.xlabel("Re")
plt.ylabel("Im")
#plt.savefig('juliaQ5.png', dpi=800) #<-----------|UNCOMMENT TO SAVE PLOTS
plt.show()
```
